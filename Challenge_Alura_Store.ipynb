# Carga de datos

### Subtarea:
Cargar los cuatro archivos CSV en los DataFrames de Pandas.

**Razonamiento**:
Cargue los cuatro archivos CSV en los DataFrames de Pandas y muestre las primeras filas de cada uno para verificar.


```python
import pandas as pd

df_loja_1 = pd.read_csv('loja_1.csv')
display(df_loja_1.head())

df_loja_2 = pd.read_csv('loja_2.csv')
display(df_loja_2.head())

df_loja_3 = pd.read_csv('loja_3.csv')
display(df_loja_3.head())

df_loja_4 = pd.read_csv('loja_4.csv')
display(df_loja_4.head())
```


#Preparación de datos

Subtarea:
Combine los cuatro DataFrames de la tienda en uno solo y agregue la columna "store_id".

**Razonamiento**:
Concatene los cuatro marcos de datos y agregue una columna 'store_id'.

```python
import pandas as pd

df_all_stores = pd.concat([df_loja_1, df_loja_2, df_loja_3, df_loja_4], ignore_index=True)

# Agregue la columna 'store_id'
store_ids = [1] * len(df_loja_1) + [2] * len(df_loja_2) + [3] * len(df_loja_3) + [4] * len(df_loja_4)
df_all_stores['store_id'] = store_ids

display(df_all_stores.head())
print(df_all_stores.shape)
```

#1. Análisis de facturación

### Subtarea: calcular los ingresos totales por tienda.


**Razonamiento**:
Calcule los ingresos totales por tienda agrupando `df_all_stores` por `store_id` y sumando la columna `Precio`.


```python
# Calcular los ingresos totales por tienda
df_revenue_per_store = df_all_stores.groupby('store_id')['Preço'].sum().reset_index()
df_revenue_per_store = df_revenue_per_store.rename(columns={'Preço': 'total_revenue'})
display(df_revenue_per_store)
```

# 2. Ventas por Categoria

### Subtarea:
Calcular el número de productos vendidos por categoría en cada tienda.

**Razonamiento**:
Agrupe el DataFrame `df_all_stores` por `store_id` y `Categoria do Produto`, conté la cantidad de productos en cada grupo usando `.size()`, restablezca el índice y cambie el nombre de la columna de recuento a 'sales_count'.

```python
# Calcular el número de productos vendidos por categoría para cada tienda
df_sales_per_category = df_all_stores.groupby(['store_id', 'Categoria do Produto']).size().reset_index(name='sales_count')
display(df_sales_per_category)
```

# 3. Calificación promedio de la tienda

### Subtarea:
Calcular la puntuación media de las reseñas de los clientes de cada tienda.

**Razonamiento**:
Calcule la puntuación media de las reseñas de los clientes para cada tienda agrupando el DataFrame `df_all_stores` por `store_id` y calculando la media de la columna 'Avaliação da compra'.

```python

# Calcular la puntuación media de las reseñas de los clientes para cada tienda
df_average_review = df_all_stores.groupby('store_id')['Avaliação da compra'].mean().reset_index()
df_average_review = df_average_review.rename(columns={'Avaliação da compra': 'average_review'})
display(df_average_review)
```

# 4. Productos más vendidos y menos vendidos

### Subtarea:
Identificar los productos más vendidos y los menos vendidos en cada tienda.

**Razonamiento**:
Agrupar los datos por tienda y nombre de producto, contar las ventas y encontrar los productos más y menos vendidos.


```python
# Agrupa los datos por store_id y nombre del producto
sales_by_product = df_all_stores.groupby(['store_id', 'Produto']).size().reset_index(name='sales_count')

# Encuentra los productos más vendidos y los menos vendidos de cada tienda
best_worst_products = []
for store in sales_by_product['store_id'].unique():
    store_data = sales_by_product[sales_by_product['store_id'] == store]
    best_product = store_data.loc[store_data['sales_count'].idxmax(), 'Produto']
    worst_product = store_data.loc[store_data['sales_count'].idxmin(), 'Produto']
    best_worst_products.append({'store_id': store, 'best_selling_product': best_product, 'worst_selling_product': worst_product})

df_best_worst = pd.DataFrame(best_worst_products)
display(df_best_worst)
```

# 5. Envío promedio por tienda

### Subtarea:
Calcular el coste medio de envío de cada tienda.

**Razonamiento**:
Calcule el costo promedio de envío por tienda y almacene el resultado en un nuevo DataFrame.

```python
# Calcula el costo de envío promedio para cada tienda
df_average_shipping = df_all_stores.groupby('store_id')['Frete'].mean().reset_index()
df_average_shipping = df_average_shipping.rename(columns={'Frete': 'average_shipping_cost'})
display(df_average_shipping)
```

## Visualización de datos

### Subtarea:
Crear visualizaciones para las métricas calculadas.

```python
import matplotlib.pyplot as plt

# 1.Ingresos totales por tienda
plt.figure(figsize=(8, 6))
plt.bar(df_revenue_per_store['store_id'], df_revenue_per_store['total_revenue'], color='skyblue')
plt.xlabel('ID de tienda')
plt.ylabel('Ingresos totales')
plt.title('Ingresos totales por tienda')
plt.savefig('gráfico_de_ingresos.png')
plt.show()
```

```python
# 2. Ventas por categoría
plt.figure(figsize=(12, 8))
stores = df_sales_per_category['store_id'].unique()
for store in stores:
    store_data = df_sales_per_category[df_sales_per_category['store_id'] == store]
    plt.bar(store_data['Categoria do Produto'], store_data['sales_count'], label=f'Store {store}')
plt.xlabel('Categoría de producto')
plt.ylabel('Recuento de ventas')
plt.title('Ventas por categoría')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig('category_chart.png')
plt.show()
```

```python
# 3. Puntuación media de las reseñas de los clientes por tienda
plt.figure(figsize=(8, 6))
plt.bar(df_average_review['store_id'], df_average_review['average_review'], color='lightcoral')
plt.xlabel('ID de tienda')
plt.ylabel('Puntuación promedio de revisión')
plt.title('Puntuación media de las reseñas de los clientes por tienda')
plt.savefig('review_chart.png')
plt.show()
```

```python
# 4. Productos más vendidos y más vendidos
for store in df_best_worst['store_id'].unique():
    store_data = df_best_worst[df_best_worst['store_id'] == store]
    products = [store_data['best_selling_product'].iloc[0], store_data['worst_selling_product'].iloc[0]]
    sales_counts = [sales_by_product[(sales_by_product['store_id'] == store) & (sales_by_product['Produto'] == store_data['best_selling_product'].iloc[0])]['sales_count'].iloc[0],
                   sales_by_product[(sales_by_product['store_id'] == store) & (sales_by_product['Produto'] == store_data['worst_selling_product'].iloc[0])]['sales_count'].iloc[0]]

    plt.figure(figsize=(8,6))
    plt.bar(products, sales_counts, color = ['green', 'red'])
    plt.xlabel('Productos')
    plt.ylabel('Recuento de ventas')
    plt.title(f'Productos más vendidos y más vendidos - Tienda {store}')
    plt.savefig(f'product_chart_store_{store}.png')
    plt.show()
```

```python
# 5. Costo de envío promedio por tienda
plt.figure(figsize=(8, 6))
plt.bar(df_average_shipping['store_id'], df_average_shipping['average_shipping_cost'], color='lightgreen')
plt.xlabel('ID de tienda')
plt.ylabel('Costo promedio de envío')
plt.title('Costo de envío promedio por tienda')
plt.savefig('shipping_chart.png')
plt.show()
```
